```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
def train_mean_std(path: str, metric="reward"):
    distances = {}
    for i, trial in enumerate(os.listdir(path)):
        data_path = os.path.join(path, trial, "train_log", "{}.txt".format(metric))
        df = pd.read_csv(data_path)
        distances[trial] = df[metric]
    all_distance_df = pd.DataFrame(distances)
    
    return all_distance_df.mean(1), all_distance_df.std(1)
```

```python
y, err = train_mean_std("results/discrepance/run0")
x = np.arange(len(y))
plt.plot(x, y, label='with discrepancy')
plt.fill_between(x, y - err, y + err, alpha=0.2)

y, err = train_mean_std("results/linear/run0")
x = np.arange(len(y))
plt.plot(x, y, linestyle='--', label='linear')
plt.fill_between(x, y - err, y + err, alpha=0.3)

plt.ylim([0.6, 2.5])

plt.legend(loc='upper left')

plt.xlabel("Episodes")
plt.ylabel("Locomotion distance per 10000 steps")

plt.savefig("graphs/LearningCurve_run0.png")
```
