```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
def mean_std(path: str, file="reward", metric="reward"):
    distances = {}
    for i, trial in enumerate(filter(lambda x: re.match(r'^trial[0-9]+$', x), os.listdir(path))):
        data_path = os.path.join(path, trial, "train_log", "{}.txt".format(file))
        df = pd.read_csv(data_path)
        distances[trial] = df[metric]
    all_distance_df = pd.DataFrame(distances)
    
    return all_distance_df.mean(1), all_distance_df.std(1)
```

```python
y, err = mean_std("results/fric_linear_tension/run1", "distance", "distance")
x = np.arange(len(y))
plt.plot(x, y, label='linear model')
plt.fill_between(x, y - err, y + err, alpha=0.2)

# y, err = train_mean_std("results/linear/run0")
# x = np.arange(len(y))
# plt.plot(x, y, linestyle='--', label='linear')
# plt.fill_between(x, y - err, y + err, alpha=0.3)

# plt.ylim([0.6, 2.5])

plt.legend(loc='lower right')

plt.xlabel("Episodes")
plt.ylabel("Locomotion distance per 10000 steps")

plt.savefig("graphs/LearningCurve_LinearFrictionTention.png")
```
