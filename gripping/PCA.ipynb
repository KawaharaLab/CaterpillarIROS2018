```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
import pickle
import os
import re
%matplotlib inline
```

```python
params = pickle.load(open("results/muscle/normal/k9/run0/trial0/model/actor_model.pickle", 'rb'))
```

```python
def bundle(model_file_path: str, params_keys: tuple):
    params = pickle.load(open(model_file_path, 'rb'))
    vectors = []
    for k in params_keys:
        v = params[k]
        vectors.append(v.flatten())
    return np.concatenate(vectors)
```

```python
# params_keys = ('b_cos', 'b_sin', 'w_cos', 'w_sin')
# params_keys = ('w_cos', 'w_sin')
# params_keys = ('b_cos', 'b_sin')
# params_keys = ('b_cos', 'w_cos')
# params_keys = ('b_sin',  'w_sin')
params_keys = ('w_sin',)

targets = []
controller_features = []

base_path = "results/muscle/normal/k3/run0/"
for trial in  filter(lambda x: re.match(r'trial[0-9]+', x), os.listdir(base_path)):
        params_path = os.path.join(base_path, trial, "model", "actor_model.pickle")
        feature = bundle(params_path, params_keys)
        targets.append("k3")
        controller_features.append(feature)
        
base_path = "results/muscle/normal/k9/run0/"
for trial in  filter(lambda x: re.match(r'trial[0-9]+', x), os.listdir(base_path)):
        params_path = os.path.join(base_path, trial, "model", "actor_model.pickle")
        feature = bundle(params_path, params_keys)
        targets.append("k9")
        controller_features.append(feature)
        
base_path = "results/muscle/normal/k15/run0/"
for trial in  filter(lambda x: re.match(r'trial[0-9]+', x), os.listdir(base_path)):
        params_path = os.path.join(base_path, trial, "model", "actor_model.pickle")
        feature = bundle(params_path, params_keys)
        targets.append("k15")
        controller_features.append(feature)
        
base_path = "results/muscle/normal/k40/run0/"
for trial in  filter(lambda x: re.match(r'trial[0-9]+', x), os.listdir(base_path)):
        params_path = os.path.join(base_path, trial, "model", "actor_model.pickle")
        feature = bundle(params_path, params_keys)
        targets.append("k40")
        controller_features.append(feature)
        
base_path = "results/muscle/normal/k100/run0/"
for trial in  filter(lambda x: re.match(r'trial[0-9]+', x), os.listdir(base_path)):
        params_path = os.path.join(base_path, trial, "model", "actor_model.pickle")
        feature = bundle(params_path, params_keys)
        targets.append("k100")
        controller_features.append(feature)
        
        
        
pca = PCA(n_components=2)
pca.fit(controller_features)
transformed = pca.fit_transform(controller_features)

targets = np.array(targets)
for label in np.unique(targets):
    plt.scatter(transformed[targets == label, 0], transformed[targets==label, 1], label=label)

plt.legend(loc='lower right')
```
