```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import MultipleLocator
%matplotlib inline
```

```python
# somite0 and somite4 on test
fontsize = 20
linestyles = ('-', ':', '-.')
colors = ('r', 'darkgreen', 'cornflowerblue')
captions = ("k = 9.0", "k = 40.0")

df_k9 = pd.read_csv("results/muscle/normal/k9/run0/trial2/run/somite_phase_diffs.txt")
df_k40 = pd.read_csv("results/muscle/normal/k40/run0/trial12/run/somite_phase_diffs.txt")

df_k9.columns = ["step"] + ["φ{}".format(i) for i in range(3)]
df_k40.columns = ["step"] + ["φ{}".format(i) for i in range(3)]

fig, axes = plt.subplots(1, 2, figsize=(19, 7), sharex=False, sharey=True)
for i, (ax, df) in enumerate(zip(axes, (df_k9, df_k40))):
    for ls, c, p in zip(linestyles, colors, df.columns[1:]):
        ax.plot(df['step'], df[p], linestyle=ls, color=c)
        
    ax.set_xlabel("Steps [x 1000]", fontsize=fontsize)
    ax.tick_params(labelsize=fontsize)
    ax.set_xticks(np.arange(0, 10001, 2500))
    ax.set_xticklabels(np.arange(0, 10.1, 2.5))
    ax.set_ylim(-np.pi*2/3, 1.1)
    ax.set_xlim(0, 10000)
    
    ax.text(4000, -2.9, captions[i], fontsize=fontsize)
    #ax.set_title(captions[i], fontsize=fontsize)

    if i == 0:
        ax.set_ylabel("Phase [rad]", fontsize=fontsize)
    
    ax.grid(which='both')
    ax.grid(which='major', linestyle='-.', linewidth='0.7', alpha=0.5)
#     ax.yaxis.set_minor_locator(MultipleLocator(np.pi / 8))
#     ax.grid(which='minor', linestyle='-.', linewidth='0.7', alpha=0.5)

ax.legend(loc='lower right', fontsize=fontsize)
plt.subplots_adjust(left=0.118, right=0.812, bottom=0.2, top=0.95, wspace=0.12)        

plt.savefig("graphs/phase_gradients_crawling_and_inching.eps")
```
